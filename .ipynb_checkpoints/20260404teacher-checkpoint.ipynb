```python
number = '0'   #변수 초기화
while number != '4':  # 반복문을 빠져나갈 조건
    print('1. 더하기')
    print('2. 빼기')
    print('3. 곱하기')
    print('4. 종료')
    number = input("번호>>")

    if number != '4':
        num1 = int(input('첫번째 숫자:'))
        num2 = int(input('두번째 숫자:')) 
    if number == '1':
        print(f"{num1} + {num2} = {num1+num2}")
    if number == '2':
        print('빼기:', num1-num2)
    if number == '3':
        print(f"{num1} X {num2} = {num1*num2}")
print('종료!!')
```

```python
x = 1   #초기화
tot = 0 #초기화
while x <= 10:
    tot += x    #  tot = tot + x   # 누적
    x += 1   # x = x + 1 1씩 증가
print(tot)
```

```python
number = "0"
prompt = """
    1. 국어 점수 입력
    2. 영어 점수 입력
    3. 수학 점수 입력
    4. 총점 및 평균 출력
    5. 종료

    번호선택>>"""
while number != "5":
    print(prompt)
    number = input()
    if number == "1":
        kor = int(input("국어점수:"))
    if number == "2":
        eng = int(input("영어점수:"))
    if number == "3":
        math = int(input("수학점수:"))
    if number == "4":
        tot = kor + eng + math;
        avg = tot / 3
        print(f"국:{kor}, 영:{eng}, 수{math}, 총점:{tot}, 평균:{avg}")
print('프로그램 종료!!')
    
```
