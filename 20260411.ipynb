```python
import random

score = []
for i in range(25):
    score.append(random.randint(0, 100))
print(score)

print('리스트를 사용해 70점 이상인 점수 출력하기')
for s in score:
    if s >= 70:
        print(score.index(s) + 1, '번', s)

print('인덱스를 사용해 70점 이상인 점수 출력하기')
for i in range(len(score)):   # 또는 for i in range(25)
    if score[i] >= 70:
        print(i + 1, '번', score[i])
```

```python
import random

score = []
for i in range(25):
    score.append(random.randint(0, 100))
print(score)

cnt = 0  # 70점 이상의 학생 수
print('리스트를 사용해 70점 이상인 학생 수 출력하기')
for s in score:
    if s >= 70:
        print(score.index(s) + 1, '번', s)
        cnt += 1
print(f"70점 이상의 학생 수 는 {cnt}명")
```

```python
todo = {}
while True:
    print("\n[메뉴]")
    print("1. 할 일 추가")
    print("2. 완료 처리")
    print("3. 전체 보기")
    print("4. 종료")
    menu = input("메뉴 선택: ")
    if menu == '1':
        task = input("할 일 입력: ")
        todo[task] = False
        print(todo)
        print("추가 완료!")
    elif menu == '2':          # ← 들여쓰기 1칸 너무 많았음
        task = input("완료할 일 입력: ")
        if task in todo:
            todo[task] = True  # ← 완료 처리 코드 빠져있었음!
            print("완료 처리!")
        else:
            print("없는 할 일입니다.")
    elif menu == '3':
        if not todo:
            print("할 일이 없습니다.")
        else:
            print("\n[TODO LIST]")
            for t, done in todo.items():
                if done:
                    status = "O"
                else:
                    status = "X"
                print(f"{t} : {status}")  # ← Print → print 소문자
    elif menu == '4':
        print("프로그램 종료")
        break
    else:
        print("잘못된 입력입니다.")
```
